### Autocorrelation, Autocovariance with the Airline Passengers Dataset

##### Step 1:  Load in the Air Passangers Dataset

```python
# Your Code Here
import pandas as pd
import numpy as np
%matplotlib inline

df= pd.read_csv('AirPassengers.csv', index_col='Month',parse_dates=True)
df.head()
```

##### Step 2: Pull the head of the dataset, and plot it

```python
# Your Code Here
df.plot(title="Air Passengers by month")
```

##### Step 3:  Calculate the Autocovariance at lag 3:

```python
# Your Code Here
X_t = df[:-3]['Passengers'] - df['Passengers'].mean()
X_t_one = df[3:]['Passengers'] - df['Passengers'].mean()
```

```python
X_covar = X_t.values * X_t_one.values
X_covar = X_covar.sum() * (1 / len(df))
X_covar
```

```python
# Your Code Here
from statsmodels.tsa.stattools import acf, acovf
acf(df)
```

```python
acovf(df)
```

##### Step 5:  Plot the Autocorrelations using the `autocorrelation_plot` in Pandas (On The Entire Dataset)

```python
# Your Code Here
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
plt.style.use('fivethirtyeight')

autocorrelation_plot(df);
```
