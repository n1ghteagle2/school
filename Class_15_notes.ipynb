```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
import pandas as pd
train= pd.read_csv('titanic/train.csv')
test=pd.read_csv('titanic/test.csv')
```

```python
import seaborn as sns

sns.catplot(x='Sex',y='Survived',kind='bar',data=train)
```

```python
sns.catplot(x='Pclass',y='Survived',kind='bar',data=train)
```

```python
X_train = train.loc[:,['Sex','Pclass']]
y_train = train['Survived']
X_test = test.loc[:,['Sex','Pclass']]
```

```python
X_train = pd.get_dummies(X_train, columns=['Sex', 'Pclass'], drop_first=True)
X_test  = pd.get_dummies(X_test, columns=['Sex', 'Pclass'], drop_first=True)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='liblinear')
logreg.get_params()
```

```python
logreg.fit(X_train, y_train)
```

```python
logreg.predict(X_test)
```

```python
logreg.predict_proba(X_test)
```

```python
logreg.predict_proba(X_test)[:, 1]
```

```python
logreg.intercept_
```

```python
logreg.coef_
```

```python
coeffs = pd.DataFrame({
    'Variable': X_train.columns,
    'Weight'  : logreg.coef_[0]
})

coeffs
```

```python
output = X_train.dot(logreg.coef_.T)+logreg.intercept_
output
```

```python
sigmoid(output)
```

```python
preds = logreg.predict(X_test)

lm_predictions = pd.DataFrame({
    'PassengerId':test.PassengerId,
    'Survived': preds
})
```

```python
lm_predictions.to_csv('titanic/lm_submissions.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.feature_importances_
```

```python
importances = pd.DataFrame({
    'Features': X_train.columns,
    'Importance': rf.feature_importances_
})

importances
```

```python
rf.predict(X_test)
```

```python
rf.predict_proba(X_test)
```

```python
# make the predictions
rf_preds = rf.predict(X_test)
```

```python
# pass them into a dataframe
rf_submissions = pd.DataFrame({
    'PassengerId': test.PassengerId,
    'Survived'   : rf_preds
})
```

```python
# make random forest submissions
rf_submissions.to_csv('titanic/rf_submissions.csv', index=False)
```
