```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('titanic.csv')
df.head(1)
```

```python
df.Sex.value_counts().plot(kind='bar',color=['Green','Black'])
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='barh',color=['Black','Red']);
```

```python
df.groupby('Sex')[['Age', 'Fare']].mean().plot(kind='bar')
```

```python
df['Age'].plot(kind='box')
```

```python
df['Embarked'].value_counts().plot(kind='pie')
```

```python
df['Age'].plot(kind='hist',bins=5)
```

```python
df.hist(column='Age',by='Embarked',layout=(1,3),figsize=(12,4),sharey=True)
```

```python
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(12, 4))
df[df.Sex == 'female'].plot('Age', 'Fare', kind='Scatter', ax=ax[0], color='red', label='Females')
df[df.Sex == 'male'].plot('Age', 'Fare', kind='Scatter', ax=ax[1], label='Males', color='blue')
```

```python
sns.catplot(x='Pclass',y='Age',data=df,kind='swarm',col='Sex')
```

```python
sns.catplot(data=df,kind='bar',x='Pclass',y='Survived',hue='Sex')
```

```python
sns.lmplot(x='Pclass',y='Survived',data=df)
```

```python
#df['Age'] = df['Age'].fillna(29)
sns.distplot(df.Age)
```

```python
sns.jointplot(x='Age',y='Fare',data=df[df['Sex'] == 'male'])
```

```python
ax = sns.distplot(df[df.Sex == 'female']['Age'],label='Females')
```

```python
sns.distplot(df[df.Sex == 'male']['Age'],label='males')
```

```python
sns.distplot(df[df.Sex == 'male']['Age'],label='males',ax=ax)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,x_vars=['Pclass','Age','Fare','SibSp'],y_vars=['Age'])
```
