```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('advertising.csv',index_col=0)
data.head()
```

```python
sm.stats.ztest(data['radio'], data['newspaper'])
```

```python
data[['radio','newspaper']].hist(sharey=True,figsize=(12,6))
```

```python
data[['radio','newspaper']].boxplot(figsize=(12,6))
```

```python
from statsmodels.graphics.gofplots import qqplot
x = qqplot(data['radio'])
```

```python
y = qqplot(data['newspaper'])
```

```python
from scipy.stats import shapiro

test_statistic, p_value = shapiro(data['radio'])
print("Shapiro test statistic: ", test_statistic, "P-vlaue: ", p_value)
```

```python
shapiro(data['radio'])
```

```python
from scipy.stats import kstest

new_data = (data['radio']-data['radio'].mean())/data['radio'].std()

kstest(new_data,cdf='norm')
```

```python
data1=[0]*50
data1.extend(np.random.normal(14,4,150))
np.random.shuffle(data1)

data2=[0]*100
data2.extend(np.random.normal(20,5,100))
np.random.shuffle(data2)

df=pd.DataFrame()
df["A"]=data1
df["B"]=data2

df.head()
```

```python
sm.stats.ztest(df["A"],df["B"])
```

```python
df[["A","B"]].boxplot(figsize=(12,6))
```

```python
from statsmodels.graphics.gofplots import qqplot
z = qqplot(df[df.A > 0]['A'])
```

```python
q = qqplot(df['B'])
```

```python
df[['A','B']].hist(sharey=True,figsize=(12,6))
```

```python
shapiro(df[['A','B']])
```

```python
new_a = (df['A']-df['A'].mean())/df['A'].std()
kstest(new_a,cdf='norm')
```

```python
new_b = (df['B']-df['B'].mean())/df['B'].std()
kstest(new_b,cdf='norm')
```

```python
from scipy.stats import ttest_ind
test_stat, p_value = ttest_ind(df['A'], df['B'], equal_var=False)
print(test_stat, p_value)
```

```python
ttest_ind(df['A'],df['B'],equal_var=True)
```
